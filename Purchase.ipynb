Analyzing the null data in the data set, it is decided to eliminate some of the variables by means of the drop function, that's because there are some columns that represent a very high percentage of nulls

The other empty values ​​it is decided to impute or fill values ​​with the average value of the data

```python
#Imput null data for '0' in the categorical variables because after numerical codes are added to the cagegoric variables
NAN_SUBSTITUTION_VALUE = "0"
data = data.astype('string')
data = data.fillna(NAN_SUBSTITUTION_VALUE)
data = data.astype('category')
```
